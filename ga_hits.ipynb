```python
import pandas as pd
```

```python
df = pd.read_csv('data/ga_hits.csv', sep = ',')
df
```

```python
values = [
    'sub_car_claim_click', 
    'sub_car_claim_submit_click', 
    'sub_open_dialog_click', 
    'sub_custom_question_submit_click', 
    'sub_call_number_click', 
    'sub_callback_submit_click', 
    'sub_submit_success', 
    'sub_car_request_submit_click'
]

# Заполнение столбца event_value
df['event_value'] = df['event_action'].apply(lambda x: 1 if x in values else 0)
df['event_value'].value_counts(dropna=False)
```

```python
# Удаляем ненужные/лишние колонки 

columns_to_drop = [
    'hit_date',
    'hit_time',
    'hit_number',
    'hit_type',
    'hit_referer',
    'hit_page_path',
    'event_category',
    'event_action',
    'event_label',
]

df.drop(columns_to_drop, axis=1, inplace=True)
df
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df = df.loc[df.groupby('session_id')['event_value'].idxmax]
df['event_value'].value_counts(dropna=False)
```
